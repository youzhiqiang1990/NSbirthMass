```python
import numpy as np

li_data_2G_pos=np.loadtxt('li_data_2G_pos.txt')
li_data_TOP_pos=np.loadtxt('li_data_TOP_pos.txt')

import corner
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.lines as mlines
%matplotlib inline
plt.style.use(['science','ieee','no-latex'])

obs_TOP_pos = np.loadtxt('obs_TOP_pos.txt')


fig= corner.corner(obs_TOP_pos, 
                       bins=40,smooth=0.9,color='#0072c1',
                       labels=[r'$\alpha$',r'$m_{\rm{min}} $',r'$ m_{\rm{max}} $',
                               r'$ \delta_m$'],
                       #quantiles=[0.16, 0.5, 0.84],
                       levels=(1 - np.exp(-0.5), 1 - np.exp(-2), 1 - np.exp(-9 / 2.)),
                       plot_density=False, plot_datapoints=True, fill_contours=True,
                       label_kwargs=dict(fontsize=20),
                       show_titles=0, title_kwargs={"fontsize": 20})
corner.corner(li_data_TOP_pos, 
                       bins=40,smooth=0.9,color='#ff7f0e',
                       labels=[r'$\alpha$',r'$m_{\rm{min}} $',r'$ m_{\rm{max}} $',
                               r'$ \delta_m$'],
                       quantiles=[0.16, 0.84],
                       levels=(1 - np.exp(-0.5), 1 - np.exp(-2), 1 - np.exp(-9 / 2.)),
                       plot_density=False, plot_datapoints=True, fill_contours=True,
                       label_kwargs=dict(fontsize=20),
                       show_titles=1, title_kwargs={"fontsize": 20},fig=fig)


blue_line = mlines.Line2D([],[],color='#0072c1', lw=2,label=r'$\rm{OBS}$')
orange_line= mlines.Line2D([],[],color='#ff7f0e',lw=2,label=r'$\rm{HF}$')

plt.legend(handles=[blue_line,orange_line],bbox_to_anchor=(1,4.2),loc='upper right',fontsize=20)#prop={'size': 8})


plt.savefig('TOP_OBS_HF.jpg')
plt.show()
```
